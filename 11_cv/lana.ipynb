```python
# First, let's install OpenCV if it's not already installed
!pip install opencv-python

# Import necessary libraries
import cv2
import numpy as np
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow  # Special function for displaying images in Colab

print("OpenCV version:", cv2.__version__)

# Since we're in Colab, we'll need to download an image to work with
!wget https://upload.wikimedia.org/wikipedia/en/7/7d/Lenna_%28test_image%29.png -O lenna.png

# Read the image
image = cv2.imread('lenna.png')

# Display the original image
print("\nOriginal Image:")
cv2_imshow(image)

# Convert to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
print("\nGrayscale Image:")
cv2_imshow(gray_image)

# Apply some image processing - Gaussian blur
blurred = cv2.GaussianBlur(image, (15, 15), 0)
print("\nBlurred Image:")
cv2_imshow(blurred)

# Edge detection using Canny
edges = cv2.Canny(gray_image, 100, 200)
print("\nEdge Detection:")
cv2_imshow(edges)

# Face detection demo (using a pre-trained Haar cascade)
# First download the Haar cascade file
!wget https://raw.githubusercontent.com/opencv/opencv/master/data/haarcascades/haarcascade_frontalface_default.xml -O haarcascade_frontalface_default.xml

# Load the cascade
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

# Detect faces
faces = face_cascade.detectMultiScale(gray_image, 1.1, 4)

# Draw rectangles around the faces
image_with_faces = image.copy()
for (x, y, w, h) in faces:
    cv2.rectangle(image_with_faces, (x, y), (x+w, y+h), (255, 0, 0), 2)

print("\nFace Detection:")
cv2_imshow(image_with_faces)

# Save processed image
cv2.imwrite('processed_lenna.png', image_with_faces)
print("\nProcessed image saved as 'processed_lenna.png'")
```
